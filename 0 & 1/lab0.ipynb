```python
# 0.6.6
def makeInverseIndex(strlist):
    word_to_ids = {}
    for doc_id, document in enumerate(strlist):
        for word in document.split():
            current_ids = word_to_ids.setdefault(word, set())
            current_ids.add(doc_id)
    
    return word_to_ids

with open(data_path + 'stories_small.txt') as f:
    text = list(f)
    inverse_index = makeInverseIndex(text)
    
inverse_index
```

```python
# 0.6.7
def orSearch(inverseIndex, query):
    doc_ids = set()
    for word in query:
        if word in inverseIndex:
            doc_ids = doc_ids | inverseIndex[word]
        
    return doc_ids

orSearch(inverse_index, ['A', 'B'])
```

```python
# 0.6.8
def andSearch(inverseIndex, query):
    doc_ids = inverseIndex[query[0]]
    for word in query[1:]:
        if word in inverseIndex:
            doc_ids = doc_ids & inverseIndex[word]
        
    return doc_ids


andSearch(inverse_index, ['prep', 'course'])
```

```python
# 0.8.3
def tuple_sum(A, B):
    return [(xA + xB, yA + yB) for (xA, yA), (xB, yB) in zip(A, B)]


tuple_sum([(1, 2), (10, 20)], [(3, 4), (30, 40)])
```
